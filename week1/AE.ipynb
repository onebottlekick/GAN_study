```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load trained autoencoder
model = tf.keras.models.load_model('models/AE.h5')
```

```python
encoder = model.layers[0]
decoder = model.layers[1]
```

```python
# load test data
(_, _), (X_test, _) = tf.keras.datasets.mnist.load_data()
```

```python
encoded =encoder.predict(np.expand_dims(X_test[0], axis=0))
decoded = decoder.predict(encoded)
```

```python
for i in range(5):
    plt.subplot(2, 5, i+1)
    plt.imshow(X_test[i], cmap='gray')
    plt.axis('off')
    plt.title('original')
    
    plt.subplot(2, 5, i+6)
    encoded =encoder.predict(np.expand_dims(X_test[i], axis=0))
    decoded = decoder.predict(encoded)
    plt.imshow(decoded.squeeze(), cmap='gray')
    plt.axis('off')
    plt.title('decoded')
```

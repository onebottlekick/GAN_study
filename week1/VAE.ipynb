```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

from VAE import VAE, encoder, decoder
```

```python
VAE.compile(optimizer='rmsprop', loss="binary_crossentropy")
(X_train, _), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
X_train = X_train/255.
X_train = X_train.reshape(-1, 784)
X_test = X_test/255.
X_test = X_test.reshape(-1, 784)
VAE.fit(X_train, X_train, shuffle=True, epochs=50, batch_size=100)
```

```python
X_test_encoded = encoder.predict(X_test, batch_size=100)[0]
plt.figure(figsize=(6, 6))
plt.scatter(X_test_encoded[:,0], X_test_encoded[:,1], c=y_test, cmap='viridis')
plt.colorbar()
plt.show()
```
